```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
data=pd.read_csv("cleaned_playStore_data.csv")
data.head()
```

```python
data[data.duplicated()]
```

```python
data=data.drop_duplicates(subset=['App'],keep='first')
```

```python
data.shape
```

```python
numeric_features = [feature for feature in data.columns if data[feature].dtype != 'O']
```

```python
numeric_features
```

```python
categorical_features = [feature for feature in data.columns if data[feature].dtype == 'O']
```

```python
categorical_features
```

```python
for col in categorical_features:
    print(col)
```

```python
for col in categorical_features:
    print(f"{col}:{data[col].value_counts(normalize=True)*100}")
    print("=============================================")
```

```python
sns.countplot(x=data["Type"])
```

```python
sns.countplot(x=data['Content Rating'])
```

```python
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=data[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```

```python
sns.kdeplot(data['Price'])
```

```python
sns.kdeplot(data['Reviews'])
```

```python
sns.distplot(data['Rating'])
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=data[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
